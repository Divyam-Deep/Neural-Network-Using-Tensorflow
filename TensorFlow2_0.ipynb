```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observation = 1000
xp = np.random.uniform(-10,10, size = (observation, 1))  #observation 1 is we want x variable in the eaquation
zp = np.random.uniform(-10,10, size = (observation, 1))
generated_input = np.column_stack((xp,zp))
noise = np.random.uniform(-1,1, size = (observation, 1))
generated_target = 2*xp - 3*zp + 5 + noise
np.savez('TF_intro', inputs = generated_input, targets = generated_target)  # this section is used for making a random inputs and then save that file in npz extension (numpy extension) as tf dosent work with csv or txt file
```

```python
training_data = np.load('TF_intro.npz')
```

```python
input_size = 2  #as there are two veriables x and z
output_size = 1

model = tf.keras.Sequential([         #keras is used because tf2.0 is based on keras and sequential is used for satck arrays (lay down)
    tf.keras.layers.Dense(output_size)  # here dense works same as we do inputs*weights+bias
])

model.compile(optimizer = 'sgd', loss = 'mean_squared_error')  #optimization algo (sgd - stochastic gradient descent) and loss function

model.fit(training_data['inputs'], training_data['targets'], epochs = 100, verbose = 2)  #verbose=0 is used to show now output in training 1 to show progress bar and 2 is used to give one line after very epochs
```

```python
model.layers[0].get_weights()
```

```python
# now lets try to predict the value with this model
model.predict_on_batch(training_data['inputs']).round(1)
```

```python
training_data['targets'].round(1)
```

```python
# we can see that our mmodel output is very close to actual values
```

```python
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])), np.squeeze(training_data['targets']))
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
